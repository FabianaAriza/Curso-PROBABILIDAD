# **Ejemplo 1: La cola al burro**
Un jugador debe colocarle la cola al burro donde ésta tiene forma circular y se encuentra dentro de un canasto cuadrangular, ¿Cual es la probabilidad de que el jugador acierte?

```python
# @title: **Ilustración**

import matplotlib.pyplot as plt
import numpy as np

# Parámetros
s = 20
r = 3

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Dibujar el cuadrado
square = plt.Rectangle((0, 0), s, s, fill=False)
ax.add_patch(square)

# Dibujar el círculo
circle = plt.Circle((s / 2, s / 2), r, color='red', fill=False)
ax.add_patch(circle)

# Configurar los límites de los ejes
ax.set_xlim(0, s)
ax.set_ylim(0, s)

# Agregar texto con las áreas
ax.text(s / 2, s * 0.9, f"Área Cuadrado = {s**2}", ha='center', va='center')
ax.text(s / 2, s * 0.1, f"Área Círculo = {np.round(np.pi * r**2, 2)}", ha='center', va='center', color='red')


# Calcular y mostrar la probabilidad
probability = np.round((np.pi * r**2) / s**2, 2)
print(f"La probabilidad de acertar es: {probability}")

# Mostrar la gráfica
plt.title("Probabilidad Geométrica: Punto en un Área")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# **2. Problema de probabilidad: Encuentro de dos amigos**

---
### **Problema**

Dos amigos acuerdan encontrarse en un lugar entre las **12:00 PM y la 1:00 PM**. Cada uno llega **en un momento aleatorio**, independientemente del otro. El que llega primero **espera 15 minutos**, y si el otro no ha llegado, **se va**. ¿Cuál es la probabilidad de que se encuentren?

---

### **Solución**

Sea:

* $X$: el tiempo (en minutos después de las 12:00 PM) en que llega el primer amigo.
* $Y$: el tiempo en que llega el segundo amigo.

Ambos llegan en un momento aleatorio entre $0$ y $60$ minutos. Esto significa que $(X, Y)$ se distribuye **uniformemente** en el cuadrado de lado $60$ unidades:

$$
0 \leq X \leq 60,\quad 0 \leq Y \leq 60
$$

Se encontrarán si **la diferencia entre sus tiempos de llegada es de 15 minutos o menos**, es decir:

$$
|X - Y| \leq 15
$$

Esto define una región en el cuadrado de lado \$60\$, entre las rectas:

* $Y = X + 15$
* $Y = X - 15$

Queremos hallar el área de esta región y dividirla entre el área total del cuadrado ($60 \times 60 = 3600$).

#### **Área de la región favorable**

La región entre $|X - Y| \leq 15$ es el cuadrado completo **menos** dos triángulos rectángulos en las esquinas donde la diferencia de tiempo es mayor a 15 minutos.

Cada uno de esos triángulos tiene catetos de longitud \$45\$ (porque \$60 - 15 = 45\$), por lo tanto:

* Área de un triángulo: $frac{1}{2} \cdot 45 \cdot 45 = 1012.5$
* Área total desfavorable: $2 \cdot 1012.5 = 2025$

Entonces, la **área favorable** es:

$$
3600 - 2025 = 1575
$$

#### **Probabilidad de encuentro**

$$
P = \frac{1575}{3600} = \frac{7}{16}
$$

---

### **Respuesta final**

La probabilidad de que los dos amigos se encuentren es:

$$
\boxed{\frac{7}{16}}
$$

### **Representación gráfica**


```python
# prompt: Escribe el anterior codigo en python

import matplotlib.pyplot as plt
import numpy as np

# Problema de probabilidad: Encuentro de dos amigos

# Creamos un rango de tiempo de 0 a 60 minutos
x = np.linspace(0, 60, 300)
y = np.linspace(0, 60, 300)

# Creamos una malla para representar todas las posibles combinaciones de tiempos de llegada
X, Y = np.meshgrid(x, y)


# Creamos una máscara para identificar los casos en los que se encuentran
# Definimos la condición de encuentro: la diferencia de tiempo es menor o igual a 15 minutos
encounter_condition = np.abs(X - Y) <= 15

# Calculamos la probabilidad
probability_meeting = np.sum(encounter_condition) / (len(x) * len(y))

print(f"\nLa probabilidad de que los dos amigos se encuentren es: {probability_meeting}")

# Representación gráfica

plt.figure(figsize=(6, 6))

# Representamos la región favorable (se encuentran)
plt.imshow(encounter_condition, extent=[0, 60, 0, 60], origin='lower', cmap='Blues', alpha=0.5)

# Dibujamos las líneas que definen la región
plt.plot(x, x + 15, linestyle='--', color='green')
plt.plot(x, x - 15, linestyle='--', color='red')

plt.xlabel("Hora de llegada del amigo A (min)")
plt.ylabel("Hora de llegada del amigo B (min)")
plt.title("Región de encuentro entre dos amigos")
plt.xlim(0,60)
plt.ylim(0,60)

plt.show()
```
